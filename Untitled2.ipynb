```python
# Validador BMP que aceita DIB 40/108/124 e imprime metadados.
# (Sem libs extras)

import struct, os
from google.colab import files

# Reutiliza o último arquivo enviado; se não existir, pede upload.
try:
    fname
except NameError:
    uploaded = files.upload()
    fname = next(iter(uploaded))

print("Arquivo:", fname, "| existe?", os.path.isfile(fname))

with open(fname, "rb") as f:
    # File header (14 bytes)
    sig = f.read(2)
    if sig != b"BM":
        raise ValueError("Não é BMP ('BM').")
    file_size, res1, res2, pixel_offset = struct.unpack("<IHHI", f.read(12))

    # DIB header size
    dib_size = struct.unpack("<I", f.read(4))[0]

    if dib_size not in (40, 108, 124):
        raise ValueError(f"DIB header {dib_size} não suportado (esperado 40, 108 ou 124).")

    # Campos comuns dos primeiros 40 bytes (BITMAPINFOHEADER)
    rest40 = f.read(36)
    width, height, planes, bpp, compression, img_size, xppm, yppm, colors, important = struct.unpack(
        "<iiHHIIiiII", rest40
    )

    # Apenas para posicionamento: pula o restante do DIB se houver (V4=108, V5=124)
    if dib_size > 40:
        f.seek(dib_size - 40, 1)  # pular bytes extras do cabeçalho

    # Não precisamos ler paleta (não usada em 24/32 bpp BI_RGB)
    # Garantir que podemos chegar ao início dos pixels
    f.seek(pixel_offset)

print(f"DIB size: {dib_size} bytes")
print(f"Dimensões: {width} x {abs(height)}  (top-down? {'sim' if height<0 else 'não, bottom-up'})")
print(f"Planos: {planes} | bpp: {bpp}")
print(f"Compressão: {compression}  (0=BI_RGB sem compressão)")
print(f"Offset dos pixels: {pixel_offset} bytes")
print(f"Tamanho declarado da imagem (pode ser 0 em BI_RGB): {img_size}")

# Regras mínimas que aceitaremos para processar:
ok = (compression == 0) and (bpp in (24, 32))
print("Compatível para processar (BI_RGB 24/32bpp)?", ok)
```

```python
# Garante fname (pede upload se necessário)
try:
    fname
except NameError:
    from google.colab import files
    uploaded = files.upload()
    fname = next(iter(uploaded))
print("Usando arquivo:", fname)

import struct

def clamp255(x):
    x = int(round(x))
    return 0 if x < 0 else (255 if x > 255 else x)

def write_pgm_p2(w, h, pixels_gray, path, maxv=255):
    with open(path, "w", encoding="utf-8") as f:
        f.write(f"P2\n{w} {h}\n{maxv}\n")
        k = 0
        for _ in range(h):
            row = pixels_gray[k:k+w]
            f.write(" ".join(str(v) for v in row) + "\n")
            k += w

def write_pbm_p1(w, h, bin_vals_0_255, out_path):
    # PBM P1: 1=preto, 0=branco (nosso bin: 0=preto, 255=branco)
    with open(out_path, "w", encoding="utf-8") as f:
        f.write(f"P1\n{w} {h}\n")
        k = 0
        for _ in range(h):
            row = bin_vals_0_255[k:k+w]
            mapped = ["1" if v == 0 else "0" for v in row]
            f.write(" ".join(mapped) + "\n")
            k += w

def read_bmp_and_make_gray(bmp_path, threshold=127):
    with open(bmp_path, "rb") as f:
        # File header
        if f.read(2) != b"BM":
            raise ValueError("Não é BMP.")
        file_size, res1, res2, pixel_offset = struct.unpack("<IHHI", f.read(12))
        # DIB header
        dib_size = struct.unpack("<I", f.read(4))[0]
        if dib_size not in (40,108,124):
            raise ValueError(f"DIB {dib_size} não suportado (preciso 40/108/124).")
        rest40 = f.read(36)
        width, height, planes, bpp, compression, img_size, xppm, yppm, colors, important = struct.unpack(
            "<iiHHIIiiII", rest40
        )
        if dib_size > 40:
            f.seek(dib_size - 40, 1)  # pular V4/V5 extra
        if compression != 0 or bpp not in (24,32):
            raise ValueError("Preciso de BI_RGB sem compressão, bpp 24 ou 32.")
        top_down = height < 0
        w = width; h = abs(height); bpp_bytes = bpp // 8
        f.seek(pixel_offset)
        row_raw = w * bpp_bytes
        padding = (4 - (row_raw % 4)) % 4

        rows = []
        for _ in range(h):
            row_data = f.read(row_raw)
            if padding: f.read(padding)
            gray_row = []
            for i in range(0, row_raw, bpp_bytes):
                B = row_data[i + 0]; G = row_data[i + 1]; R = row_data[i + 2]
                y = 0.299*R + 0.587*G + 0.114*B
                gray_row.append(clamp255(y))
            rows.append(gray_row)
        if not top_down:
            rows.reverse()
        grays = [v for row in rows for v in row]
        bins = [255 if v >= threshold else 0 for v in grays]
        return w, h, grays, bins, bpp, top_down, padding

# Executa
out_gray     = "gray_from_bmp.pgm"
out_bin_pgm  = "binary_from_bmp.pgm"
out_bin_pbm  = "binary_from_bmp.pbm"
thr = 127

w, h, grays, bins, bpp_used, top_down, pad = read_bmp_and_make_gray(fname, threshold=thr)
write_pgm_p2(w, h, grays, out_gray, 255)
write_pgm_p2(w, h, bins,  out_bin_pgm, 255)
write_pbm_p1(w, h, bins,  out_bin_pbm)

print(f"OK! Gerados:\n - {out_gray}\n - {out_bin_pgm}\n - {out_bin_pbm}")
print(f"Dimensões: {w}x{h} | bpp={bpp_used} | top_down={top_down} | padding/linha={pad} bytes")

# Prévia das primeiras linhas do PGM cinza
with open(out_gray, "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        print(line.rstrip())
        if i >= 7:
            break
```

```python
# Mostrar: BMP original (RGB), grayscale PGM, binária PGM e binária PBM
# Versão corrigida: usa np.frombuffer + reshape para decodificar linhas do BMP

import os, struct
import numpy as np
import matplotlib.pyplot as plt

def read_bmp_to_rgb_np(path):
    with open(path, "rb") as f:
        if f.read(2) != b"BM":
            raise ValueError("Não é BMP.")
        file_size, r1, r2, pixel_offset = struct.unpack("<IHHI", f.read(12))
        dib_size = struct.unpack("<I", f.read(4))[0]
        if dib_size not in (40,108,124):
            raise ValueError(f"DIB {dib_size} não suportado (40/108/124).")
        width, height, planes, bpp, compression, img_size, xppm, yppm, colors, important = struct.unpack(
            "<iiHHIIiiII", f.read(36)
        )
        if dib_size > 40:
            f.seek(dib_size - 40, 1)  # pular V4/V5 extra
        if compression != 0 or bpp not in (24,32):
            raise ValueError("Preciso de BI_RGB sem compressão, bpp 24 ou 32.")
        top_down = height < 0
        w, h = width, abs(height)
        bpp_bytes = bpp // 8
        f.seek(pixel_offset)
        row_raw = w * bpp_bytes
        padding = (4 - (row_raw % 4)) % 4

        rows = []
        for _ in range(h):
            row = f.read(row_raw)
            if len(row) != row_raw:
                raise ValueError("Linha de pixels truncada.")
            if padding:
                f.read(padding)  # descarta padding de alinhamento
            arr = np.frombuffer(row, dtype=np.uint8).reshape(w, bpp_bytes)
            # BMP = B,G,R,(A). Monta RGB
            rgb = np.empty((w, 3), dtype=np.uint8)
            rgb[:, 0] = arr[:, 2]  # R
            rgb[:, 1] = arr[:, 1]  # G
            rgb[:, 2] = arr[:, 0]  # B
            rows.append(rgb)
        if not top_down:
            rows.reverse()
        img = np.stack(rows, axis=0)  # (h, w, 3)
        return img

def read_pgm_p2_to_np(path):
    tokens = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith("#"): continue
            tokens.extend(line.split())
    assert tokens[0] == "P2", "Arquivo não é PGM P2."
    i = 1
    w = int(tokens[i]); i += 1
    h = int(tokens[i]); i += 1
    maxv = int(tokens[i]); i += 1
    vals = np.array(list(map(int, tokens[i:i+w*h])), dtype=np.float32).reshape(h, w)
    if maxv != 255 and maxv > 0:
        vals *= (255.0/maxv)
    return vals.astype(np.uint8)

def read_pbm_p1_to_np(path):
    tokens = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith("#"): continue
            tokens.extend(line.split())
    assert tokens[0] == "P1", "Arquivo não é PBM P1."
    i = 1
    w = int(tokens[i]); i += 1
    h = int(tokens[i]); i += 1
    bits = np.array(list(map(int, tokens[i:i+w*h])), dtype=np.uint8).reshape(h, w)
    # PBM: 1=preto, 0=branco -> mapeia para 0..255
    return np.where(bits == 1, 0, 255).astype(np.uint8)

# usa 'fname' já definido; se não existir, pede upload
try:
    fname
except NameError:
    from google.colab import files
    uploaded = files.upload()
    fname = next(iter(uploaded))

rgb = read_bmp_to_rgb_np(fname)  # (h,w,3)

# grayscale: do arquivo se existir; senão calcula do BMP
if os.path.exists("gray_from_bmp.pgm"):
    gray = read_pgm_p2_to_np("gray_from_bmp.pgm")
else:
    gray = (0.299*rgb[...,0] + 0.587*rgb[...,1] + 0.114*rgb[...,2]).round().astype(np.uint8)

# binário PGM: do arquivo se existir; senão threshold=127
if os.path.exists("binary_from_bmp.pgm"):
    binary_pgm = read_pgm_p2_to_np("binary_from_bmp.pgm")
else:
    binary_pgm = np.where(gray >= 127, 255, 0).astype(np.uint8)

# binário PBM: do arquivo se existir; senão usa o binário PGM
if os.path.exists("binary_from_bmp.pbm"):
    binary_pbm = read_pbm_p1_to_np("binary_from_bmp.pbm")
else:
    binary_pbm = binary_pgm.copy()

fig, axs = plt.subplots(1, 4, figsize=(16, 4))
axs[0].imshow(rgb);                      axs[0].set_title("BMP original"); axs[0].axis("off")
axs[1].imshow(gray, cmap="gray", vmin=0, vmax=255);   axs[1].set_title("Grayscale (PGM)"); axs[1].axis("off")
axs[2].imshow(binary_pgm, cmap="gray", vmin=0, vmax=255); axs[2].set_title("Binary (PGM)"); axs[2].axis("off")
axs[3].imshow(binary_pbm, cmap="gray", vmin=0, vmax=255); axs[3].set_title("Binary (PBM)"); axs[3].axis("off")
plt.tight_layout(); plt.show()
```
